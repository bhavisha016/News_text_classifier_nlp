```python
import pandas as pd
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv(r"D:\csv files\News Categoires.csv",encoding='latin1')
df.head()
```

```python
le=LabelEncoder()
df["new_category"]=le.fit_transform(df.Category)
df.head()
```

```python
#preprocess
nlp=spacy.load("en_core_web_sm")
def preprocess(text):
    doc=nlp(text)
    filtered=[]
    for token in doc:
        if not  token.is_stop and not  token.is_punct:
           filtered.append(token.lemma_) 
    return " ".join(filtered)
        
    
```

```python
print(len(df["News"][0]))
```

```python
preprocess(df['News'][0])
```

```python
v=CountVectorizer()
```

```python
x=[preprocess(text) for text in df["News"]]## preprocessing all the texts in df.news
x[:8]
```

```python
x_label=v.fit_transform(x)##changong preprocessed text in vector
x_label.toarray()
```

```python
y_label=df.new_category
y_label
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_label,y_label,train_size=0.8)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
